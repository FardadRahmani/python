### Variable Funktionsparameter 端bergeben

```python
def f(**args):
    print(args)
    
f(key="value", key2="Value 2")
```

```python
def g(key, param2):
    print(key)
    print(param2)
    
d = {"key": "Ich bin der Schl端ssel", "param2": "Ich bin der Parameter"}

g(key=d["key"], param2=d["param2"])
#einfacher:
g(**d)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

def create_plot(**plot_params):#ich nehme parameter als Dictionary entgegen
    print(plot_params)#beweis dass Dictionary
    
    plt.plot([1, 2, 3], [5, 6, 5], **plot_params)
    plt.show()
    
create_plot(color="r", linewidth=10, linestyle="dashed")#ich 端bergebe beliebig viele parameter
```
