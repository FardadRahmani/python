### Idee: Wir ignorieren konkrete Laufzeiten (Vereinfachung)

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

plt.plot(np.arange(1, 600), np.arange(1, 600), label="f(n) = n")
plt.plot(np.arange(1, 600), np.arange(1, 600) * 2, label="f(n) = 2n") #diese steigt natürlich stärker an als obige

                                            #Wahre Aussage bei beiden obigen Programmen: Verdoppelung von N -->Verdoppelung Laufzeit!!
                                            #Asymptotisch gesehen haben beide also die gleiche Laufzeit, obwohl Graph anders aussieht!!
plt.legend()

plt.xlabel("n")
plt.ylabel("Anzahl print() - Befehle")
plt.show()
```

```python
np.arange(1, 6)
```

```python

plt.plot(np.arange(1, 6), np.repeat(100, 5), label="f(n) = 100")
plt.plot(np.arange(1, 6), np.arange(1, 6), label="f(n) = n")
plt.legend()

plt.xlabel("n")
plt.show()
```

```python
np.repeat(100, 5)
```

```python
# Konstante Laufzeit vs. Lineare Laufzeit

plt.plot(np.arange(1, 6), np.repeat(100, 5), label="f(n) = 100")
plt.plot(np.arange(1, 6), np.arange(1, 6), label="f(n) = n")
plt.legend()

plt.xlabel("n")
plt.show()           #Frage: welche ist zu bevorzugen? Graph sagt ja linear...
```

```python
# siehe: beei Erhöhung der Zahl zeigt sich der Unterschied klar!!

plt.plot(np.arange(1, 600), np.repeat(100, 599), label="f(n) = 100")
plt.plot(np.arange(1, 600), np.arange(1, 600), label="f(n) = n")
plt.legend()

plt.xlabel("n")
plt.show()
```

```python
# Konstante Laufzeit vs. Lineare Laufzeit

plt.plot(np.arange(1, 600), np.repeat(1, 599), label="f(n) = 1")
plt.plot(np.arange(1, 600), np.arange(1, 600), label="f(n) = n")
plt.legend()

plt.xlabel("n")
plt.show()

            #Aussage: Blaues Programm: konstante Laufzeit
            #Laufzeit orangenes Programm: wächst mit zunehmender Problemgröße!

            #wie setzen also konkrete Laufzeiten auf Null, so kann man bessere Aussagen treffen und leichter rechnen
```
