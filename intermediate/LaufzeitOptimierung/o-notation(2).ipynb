Um zu verdeutlichen, dass wir die konkrete Laufzeit weggelassen haben, packen wir das ganze in die O-Notation:

### Laufzeit: O(1)

```python
#Beispiel mit Fkt mit konstanter Laufzeit: (egal welcher Wert n hat, läuft es dreimal durch:

### Laufzeit: O(3) kann man vereinfachen denn konkretes (printZahl) interessiert uns nicht: O(1)-->Laufzeit hat nichts mit n zutun!!
def f(n):
    print("Hallo Welt")
    print("Hallo Welt")
    print("Hallo Welt")
    
f(10)
```

### Laufzeit: O(n)

```python
def f(n):
    for i in range(0, n):
        print("Hallo Welt")
        print("Hallo Welt")
                            #Dieses Bsp wäre auch dann O(n) wenn 100 print Befehler da stünden! es sei denn es kommt noch eine schleife zu
f(4)
```

### Laufzeit: O(n)

```python
def f(n):
    print("Hallo Welt")
    print("Hallo Welt")
    for i in range(0, n):
        print("Hallo Welt")
                                ##Auch hier genau gleich, die 2 print befehle interessieren nicht! immer noch O(n)
f(4)
```

## Laufzeit O(n^2)

```python
def f(n):
    for i in range(0, n): # O(n*n)
        for j in range(0, n): # O(n)
            print("Hallo Welt") # O(1)
    
f(4)
```

### Laufzeit: O(n^2)

```python
def f(n):
    for i in range(0, int(n / 2)): # O(n*n)
        for j in range(0, int(n / 2)): # O(n) 
            print("Hallo Welt") # O(1)
    
f(4)
```
