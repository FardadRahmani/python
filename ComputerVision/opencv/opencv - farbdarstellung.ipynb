```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
img = cv2.imread("bild.jpg")
#CV ist in c++ geschrieben und stellt bild in BGR (Blau grn rot) dar statt RGB
print(img[0][0])

plt.imshow(img)#matplot liest das Bild aber in RGB, daher falsche Darstellung des Bilds
plt.show()
```

```python
#CV ist in c++ geschrieben und stellt bild in BGR (Blau grn rot) dar statt RGB
#matplot liest das Bild aber in RGB, daher falsche Darstellung des Bilds
#Lsung:

print(img[0][0])
i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


print(i[0][0])


plt.imshow(i)
plt.show()
```

```python
#Bild kann auch in Graustufen konvertiert werden:

graustufen = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(graustufen, "gray")
plt.show()
```
