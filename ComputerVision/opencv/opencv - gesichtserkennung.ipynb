```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread("bild.jpg")
```

```python
i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, "gray")
plt.show()
```

### Gesichtserkenner: CLASSIFIER

```python
classifier = cv2.CascadeClassifier("./haarcascade_frontalface_alt2.xml")
#lade dir deine antrainierten Daten! (der Classifier ist trainiert anhand Millionen Bildern ein Gesicht zu erkennen)
```

```python
faces = classifier.detectMultiScale(gray, minNeighbors=1) #, minNeighbors=10 überprüft doppelt ob tatsächlich Gesicht
#durch verschiebung der Box, bei 1mal verschieben nur höhere Fehleranfälligkeit
#siehe: 2 Gesichter wurden erkannt!
```

```python
print(faces)
```

```python
c = img.copy()
for face in faces:
    x, y, w, h = face
    cv2.rectangle(c, (x, y), (x + w, y + h), (0, 255, 0), 2)
    print(face)

i = cv2.cvtColor(c, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```

```python
faces = classifier.detectMultiScale(gray, minNeighbors=10) #, minNeighbors=10 überprüft doppelt ob tatsächlich Gesicht
#durch verschiebung der Box, bei 10 mal verschieben  kleine Fehleranfälligkeit



c = img.copy()
for face in faces:
    x, y, w, h = face
    cv2.rectangle(c, (x, y), (x + w, y + h), (0, 255, 0), 2)
    print(face)

i = cv2.cvtColor(c, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```
