```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import cv2
```

### Rechteck einzeichnen:

```python
img = cv2.imread("bild.jpg")

cv2.rectangle(img, (70, 10), (230, 200), (0, 0, 255), 5)

i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```

### Kreis einzeichnen:

```python
img = cv2.imread("bild.jpg")

cv2.circle(img, (150, 100), 90, (0, 255, 0), 5) #(150, 100)->Mittelpunkt; 90->Radius

i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```

### Linie einzeichnen:

```python
img = cv2.imread("bild.jpg")

cv2.line(img, (100, 100), (200, 100), (255, 0, 0), 5) #(150, 100)->Mittelpunkt; 90->Radius

i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```

### Ellipse einzeichnen:

```python
img = cv2.imread("bild.jpg")

cv2.ellipse(img, (150, 100), (100, 50), 0, 0, 180, (0, 255, 0), 5)#(150, 100)->Mittelpunkt; (100, 50)->Lngen der Haupt- und Nebenachse
# 0 ist der Rotationswinkel der Ellipse; 0 und 180 sind die Start- und Endwinkel des Bogens
i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```

```python
img = cv2.imread("bild.jpg")

cv2.ellipse(img, (150, 100), (100, 50), 90, 0, 180, (0, 255, 0), 5)
cv2.ellipse(img, (150, 100), (100, 50), 270, 0, 180, (0, 255, 0), 5)
i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```
