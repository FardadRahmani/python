## OpenCV: Helligkeit verändern

In dieser Lektion lernst du:

- Wie das Bild intern gespeichert wird
- Wie du die Helligkeit vom Bild verändern kann ohne overflownst

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread("bild.jpg")
```

```python
print(img.shape)
```

```python
#Idee. ein Array erstellen mit genau derselben Dimension wie unser Bild:
z = np.zeros((427, 321, 3), dtype="uint8") + 50 # erstelle ein array von nullen mit Dimensionen: 427, 321, 3

print(z)
```

```python
increased = cv2.add(img, z) #Helligkeit erhöhen durch add-Methode von cv2
```

```python
print(img[200][166])
print(z[200][166])
print(increased[200][166])  #kein Overflow, durch add-methode!
```

```python
i = cv2.cvtColor(increased, cv2.COLOR_BGR2RGB)


plt.imshow(i)
plt.show()
```

```python
#Originalbild:

i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


plt.imshow(i)
plt.show()
```
